```python
#Detecting outliers using z-score and IQR
import pandas as pd
import numpy as np

#creating dataframe
data = {'column_name': [10, 15, 12, 8, 20,7,300000]}
df = pd.DataFrame(data)

#calculate z scores
z_scores = (df['column_name']-df['column_name'].mean)/df['column_name'].std()

#define threshold for outlier detector
threshold = 3

#identify outlier based on threshold
outliers = df[abs(z_scores)>threshold]

#checking for outliers
if outliers.empty:
    print("No outliers")
else:
    print(outliers)
```

```python
import pandas as pd
import numpy as np

data = {'column_name': [10,15,12,8,20,25,7,30000]}
df = pd.DataFrame(data)

#calculate first quartile Q1
Q1 = df['column_name'].quantile(0.25)

#third Q3
Q3 = df['column_name'].quantile(0.75)

#inter quartile range
IQR = Q3 - Q1

#defining threshold for outlier detection
lower_thresh = Q1-1.5*IQR
upper_thresh = Q3+1.5*IQR

#identifying outliers based on threshold
outliers = df[(df['column_name']<lower_thresh) | (df['column_name']>upper_thresh)]

if outliers.empty:
    print("No outliers")
else:
    print(outliers)
```
