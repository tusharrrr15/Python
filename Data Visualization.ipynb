Different kind of plots

Line plots

```python
import pandas as pd
import matplotlib.pyplot as plt

#create a dataframe

data = {'year':[2010,2011,2012,2013,2014,2015],
       'sales': [100,150,120,180,200,250]}

df = pd.DataFrame(data)
df

#create a line plot
plt.plot(df['year'],df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the years')

#display plot
plt.show()
```

```python
#mulitple lines
data = {'year':[2010,2011,2012,2013,2014,2015],
       'sales1': [100,150,120,180,200,250],
       'sales2':[80,120,110,160,190,220]}

df = pd.DataFrame(data)
df

#create a line plot
plt.plot(df['year'],df['sales1'],label='Sales1')
plt.plot(df['year'],df['sales2'],label='Sales2')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the years')
plt.legend()

#display plot
plt.show()
```

Bar plots

```python
data = {'year':[2010,2011,2012,2013,2014,2015],
       'sales': [100,150,120,180,200,250]}

df = pd.DataFrame(data)

#create a bar plot

plt.bar(df['year'],df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales by Year')

#display plot
plt.show()
```

Scatter plots

```python
data = {'x':[1,2,3,4,5],
       'y':[2,4,6,8,10]}

df = pd.DataFrame(data)

#create a scatter plot
plt.scatter(df['x'],df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot')

#display
plt.show()

```

```python
#with a trend line
import seaborn as sns


#create a scatter plot with a trend line
sns.regplot(x='x', y='y', data=df)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot with a trend line')

#display
plt.show()
```

Histogram

```python
import numpy as np

#generate random data

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

#create a dataframe
df = pd.DataFrame({'values': data})

#create a histogram
plt.hist(df['values'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('HISTOGRAM')

#display
plt.show()
```

Box plots


```python
data={'category': ['A','A','A','B','B','B','B'],
     'value': [10,15,20,25,30,35,40]}

df=pd.DataFrame(data)
df

sns.boxplot(x='category', y='value',data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('BOXPLOT')

#display
plt.show()
```

Heatmap

```python
data={'A':[1,2,3],
     'B':[4,5,6],
     'C':[7,8,9]}

df = pd.DataFrame(data)

sns.heatmap(df, annot=True, cmap='Blues')
plt.xlabel('Column')
plt.ylabel('Rows')
plt.title('HEATMAP')

#display
plt.show()
```
